```python
#Hacemos una lista de los archivos con terminacion .xlsm
archivos=[file for file in os.listdir() if ".xlsm" in file]
```

```python
df2= pd.DataFrame()
df2['Entidad'] = None
df2 = df2.assign(Módulo=None, Si=None, No=None, Total=None, PorcentajeCumplimientoOctubre=None, PorcentajeTrimestral=None, )

for archivo in archivos:
  df=pd.read_excel(archivo)
  M1=df.iloc[11:49, 6:15] 
  M1si=(M1=="si").sum().sum()
  M1no=(M1=="no").sum().sum()
  M1total=M1si+M1no
  M1porcerntaje= M1si /(M1si + M1no)*100 

  M1oct=df.iloc[11:49,6:9]
  M1octsi=(M1oct=="si").sum().sum()
  M1octno=(M1oct=="no").sum().sum()
  M1octtotal=M1octsi+M1octno
  M1octporcentaje= M1octsi /(M1octsi + M1octno)*100 




  filaM1= {"Entidad": (str(archivo)[:-5])[2:], 'Módulo': 1, 'Si': M1si, 'No': M1no , 'Total': M1total , 'PorcentajeTrimestral' : M1porcerntaje, 'PorcentajeCumplimientoOctubre' : M1octporcentaje } # creamos un diccionario
  df2 = df2.append(filaM1, ignore_index=True)


  M2=df.iloc[58:81, 6:15]
  M2si=(M2=="si").sum().sum()
  M2no=(M2=="no").sum().sum()
  M2total=M2si+M2no
  M2porcerntaje= M2si /(M2si + M2no)*100 

  M2oct=df.iloc[58:81,6:9]
  M2octsi=(M2oct=="si").sum().sum()
  M2octno=(M2oct=="no").sum().sum()
  M2octtotal=M2octsi+M2octno
  M2octporcentaje= M2octsi /(M2octsi + M2octno)*100 

  filaM2= { "Entidad": (str(archivo)[:-5])[2:], 'Módulo': 2, 'Si': M2si, 'No': M2no , 'Total': M2total , 'PorcentajeTrimestral' : M2porcerntaje , 'PorcentajeCumplimientoOctubre' : M2octporcentaje} # creamos un diccionario
  df2 = df2.append(filaM2, ignore_index=True)


  M3=df.iloc[89:98, 6:15]
  M3si=(M3=="si").sum().sum()
  M3no=(M3=="no").sum().sum()
  M3total=M3si+M3no
  M3porcerntaje= M3si /(M3si + M3no)*100 

  


  filaM3= {"Entidad": (str(archivo)[:-5])[2:] ,'Módulo': 3, 'Si': M3si, 'No': M3no , 'Total': M3total , 'PorcentajeTrimestral' : M3porcerntaje} # creamos un diccionario
  df2 = df2.append(filaM3, ignore_index=True)

  M4=df.iloc[106:128, 6:15]
  M4si=(M4=="si").sum().sum()
  M4no=(M4=="no").sum().sum()
  M4total=M4si+M4no
  M4porcerntaje= M4si /(M4si + M4no)*100 
  filaM4= {  "Entidad": (str(archivo)[:-5])[2:], 'Módulo': 4, 'Si': M4si, 'No': M4no , 'Total': M4total , 'PorcentajeTrimestral' : M4porcerntaje} # creamos un diccionario
  df2 = df2.append(filaM4, ignore_index=True)
  
print(df2)
 
```
